```python
athletes = spark.read.format("csv").option("header","true").option("inferSchema","true").load("Files/raw/Athletes.csv")
coaches = spark.read.format("csv").option("header","true").option("inferSchema","true").load("Files/raw/Coaches.csv")
entriesgender = spark.read.format("csv").option("header","true").option("inferSchema", "true").load("Files/raw/EntriesGender.csv")
medals = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("Files/raw/Medals.csv")
teams = spark.read.format("csv").option("header","true").option("inferSchema", "true").load("Files/raw/Teams.csv")

#display(df)
```

```python
athletes.show()
athletes.printSchema()
display(athletes)
```

```python
coaches.show()
coaches.printSchema()
display(coaches)
```

```python
entriesgender.show()
entriesgender.printSchema()
display(entriesgender)
```

```python
from pyspark.sql.functions import col
from pyspark.sql.types import IntegerType, DoubleType, BooleanType, DateType
```

```python
entriesgender = entriesgender.withColumn("Female", col("Female").cast(IntegerType()))
entriesgender = entriesgender.withColumn("Male", col("Male").cast(IntegerType()))
entriesgender = entriesgender.withColumn("Total", col("Total").cast(IntegerType()))
```

```python
entriesgender.printSchema()
```

```python
medals = medals.withColumnRenamed("Rank by Total", "Rank_by_Total")
```

```python
medals.show()
medals.printSchema()
display(medals)
```

```python
teams.show()
teams.printSchema()
display(teams)
```

```python
# Find the top countries with the highest number of gold medals
top_gold_medal_countries = medals.orderBy("Gold", ascending=False).select("Team_Country","Gold").show()
```

```python
athletes.repartition(1).write.mode("overwrite").option("header",'true').csv("Files/transformed/athletes")
coaches.repartition(1).write.mode("overwrite").option("header","true").csv("Files/transformed/coaches")
entriesgender.repartition(1).write.mode("overwrite").option("header","true").csv("Files/transformed/entriesgender")
medals.repartition(1).write.mode("overwrite").option("header","true").csv("Files/transformed/medals")
teams.repartition(1).write.mode("overwrite").option("header","true").csv("Files/transformed/teams")
```
